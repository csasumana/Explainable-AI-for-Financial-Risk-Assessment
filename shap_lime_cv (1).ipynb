```python
def agg_numeric(df, group_var, df_name):
    df = df.select_dtypes(include=[np.number])
    agg_funcs = ['mean', 'median', 'max', 'min', 'std']
    agg = df.groupby(group_var).agg(agg_funcs)
    agg.columns = pd.Index([f'{df_name}_{col[0]}_{col[1].upper()}' for col in agg.columns.tolist()])
    return agg.reset_index()
```

```python
bureau_agg = agg_numeric(bureau, 'SK_ID_CURR', 'BUR')
previous_app_agg = agg_numeric(previous_app, 'SK_ID_CURR', 'PREV')
installments_agg = agg_numeric(installments, 'SK_ID_CURR', 'INST')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate

def evaluate_model(model, X, y):
    skf = StratifiedKFold(n_splits=5)
    scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']

    scores = cross_validate(model, X, y, cv=skf, scoring=scoring, n_jobs=-1)

    results = {metric: np.mean(scores['test_' + metric]) for metric in scoring}
    return results

```
